```python
import sqlite3
conn = sqlite3.connect("event_planning.db")
cursor = conn.cursor()
```

```python
cursor.execute("SELECT name FROM sqlite_master;")
tables = cursor.fetchall()
print(tables)
```

```python
cursor.execute("SELECT * FROM moderators;")
moderators = cursor.fetchall()
print(moderators)
```

```python

cursor.execute("PRAGMA table_info(moderators);")
moderators_table_info = cursor.fetchall()
print(moderators_table_info)

# Moderators Fields
#(id,name,city,description,email,phone,expertise,created_at)
```

```python
cursor.execute("SELECT * FROM participants;")
participants = cursor.fetchall()
print(participants)
```

```python

cursor.execute("PRAGMA table_info(participants);")
participants_table_info = cursor.fetchall()
print(participants_table_info)

# Participants Fields
#(id,name,email,company,role,phone,created_at)
```
